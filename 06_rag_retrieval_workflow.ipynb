```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_ollama import ChatOllama

llm = ChatOllama(model = "Qwen2.5:latest")
```

```python

system_message = """You are a helpful HR Assistant, Answer the question only from below context"""
human_message = """{user_question} + {context}"""
```

```python
template = ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template(system_message),
    HumanMessagePromptTemplate.from_template(f"{context} + {user_question}")
])
```

```python
embeddings = HuggingFaceEmbeddings(
    model="sentence-transformers/all-MiniLM-L6-v2"
)
vectorstore_from_dir = Chroma(persist_directory = "./vectordb",
                              embedding_function = embeddings)
```

```python
user_question = "What are your working hour policy?"
retrieved_documents = vectorstore_from_dir.similarity_search(query = user_question,
                                       k = 2)
```

```python
context = " "
for doc in retrieved_documents:
    context += doc.page_content
print(context)
```

```python
message = template.format(context=context, user_question = user_question)
print(message)
```

```python
hr_assistant_response = llm.invoke(message)
print(hr_assistant_response.content)
```
