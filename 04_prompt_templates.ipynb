# 1. Initializing LLM

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key = "Enter your own api key"
    model="meta-llama/Llama-3.1-8B-Instruct:cerebras",
    max_tokens=500
)
```

```python
from langchain_ollama import ChatOllama

llm  = ChatOllama(model = "Qwen2.5:latest")
system_message = """You are a helpful travel planner."""
human_message = "Plan a travel to {destination} for {number_of_days}"
```

# 2. Define Agent Templates with variables

```python
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate
template = ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template(system_message),
    HumanMessagePromptTemplate.from_template(f"{human_message}")
])
```

# 3. Take Input from the User.

```python
print("I am your travel planner, I need few details before I can plan an itenary for you")

destination = input("What would be your destination?")

number_of_days = input(f"How long you are planning to stay in {destination}")
```

# 4. Format template variables

```python
message = template.format(destination=destination,number_of_days=number_of_days)
```

```python
response = llm.invoke(message)
print(response.content)
```
