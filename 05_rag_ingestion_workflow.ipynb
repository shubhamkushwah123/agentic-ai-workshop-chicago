# 2. Importing Libraries

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters.character import CharacterTextSplitter
```

# 3. Loading PDF Document

```python
doc_to_load = PyPDFLoader("HR_Policy_RAG.pdf")
hr_policy_doc = doc_to_load.load()
len(hr_policy_doc)
```

# 4. Splitting Document into Small Chunks

```python
chunk_splitter = CharacterTextSplitter(separator = "",
                                      chunk_size = 500,
                                      chunk_overlap = 50)
hr_policy_chunks = chunk_splitter.split_documents(hr_policy_doc)
len(hr_policy_chunks)
```

# 5. Defining Embedding Models (Convert text into vectors)

```python
from langchain_huggingface import HuggingFaceEmbeddings
# Free local embedding model
embeddings = HuggingFaceEmbeddings(
    model="sentence-transformers/all-MiniLM-L6-v2"
)
```

# 6. Generate embeddings and Store them into vector store

```python
from langchain_chroma import Chroma
vectorstore = Chroma.from_documents(documents = hr_policy_chunks,
                                    embedding = embeddings,
                                    persist_directory = "./vectordb")
print(vectorstore)
```

# 7. Retrieving relavant chunks as Context

```python
query = "What is your leave policy?"
retrieved_documents = vectorstore.similarity_search(query = query, k = 2)
for document in retrieved_documents:
    print(document.page_content)
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters.character import CharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma

# Loading Document
doc_to_load = PyPDFLoader("HR_Policy_RAG.pdf")
hr_policy_doc = doc_to_load.load()

# Initialize Splitter
chunk_splitter = CharacterTextSplitter(separator = "",
                                      chunk_size = 500,
                                      chunk_overlap = 50)
#Splitting the Document into Chunks
hr_policy_chunks = chunk_splitter.split_documents(hr_policy_doc)


# Initialiing Free embedding model from Huggingface
embeddings = HuggingFaceEmbeddings(
    model="sentence-transformers/all-MiniLM-L6-v2"
)

# Storing the embedding into vectorstore
vectorstore = Chroma.from_documents(documents = hr_policy_chunks,
                                    embedding = embeddings,
                                    persist_directory = "./vectordb")

# Querying the vector store
query = "What is your working hour policy?"
retrieved_documents = vectorstore.similarity_search(query = query, k = 2)

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings

llm = ChatOpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key="Enter your Key here ",
    model="meta-llama/Llama-3.1-8B-Instruct:cerebras",
    max_tokens=500
)
for document in retrieved_documents:
   print(document.page_content)
```
