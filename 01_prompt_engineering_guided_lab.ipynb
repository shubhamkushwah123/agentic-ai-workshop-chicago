
## 🎭 Lab 5: Role Prompting

### Objective:
Control tone and expertise.

### Task:
Change the role and stream outputs and limit the output to 300 words.


```python
prompt = '''
You are a senior DevOps engineer, answer in max 300 words.

Explain CI/CD pipelines.
'''

streams = llm.stream(prompt)
for stream in streams:
    print(stream.content, end="")
#print(response.content)
```


## 🧠 Lab 6: Chain-of-Thought

### Objective:
Encourage step-by-step reasoning.

### Task:

```python
prompt = '''
Solve step by step:

If a car travels 80 km in 1 hour, how far in 2.5 hours?
'''

response = llm.invoke(prompt)
print(response.content)
```


## 📏 Lab 7: Output Formatting


## 🚫 Lab 8: Constraints

### Objective:


## 🔁 Lab 9: Iteration

### Objective:
Compare weak vs strong prompts.

### Task:
Run and compare outputs.


```python
prompts = [
    "Explain AI",
    "Explain AI in 5 bullet points for beginners with examples"
]

for p in prompts:
    print("\nPROMPT:", p)
    response = llm.invoke(p)
    print(response.content)
```
